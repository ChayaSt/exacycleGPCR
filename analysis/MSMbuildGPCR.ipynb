```python
%matplotlib inline
from matplotlib.pyplot import * 
import mdtraj as md
import numpy as np
import pandas as pd
import os
from glob import glob
from mixtape.cluster import KCenters
from mixtape.markovstatemodel import MarkovStateModel
from mixtape import utils
from sklearn.pipeline import Pipeline
import fastcluster
```

```python
APO_B2AR = glob(os.path.expanduser('/Volumes/MyPassport/exacycleGPCR/dcd_trajectories/apo_b2ar/trj110*'))
PDB=('/Volumes/MyPassport/exacycleGPCR/GPCR_NatureChemistry/reference-structures/apo_snapshot.pdb')

tlist = []
for traj in APO_B2AR:
    tlist.append(md.load(traj, top=PDB))

trajlist = []
for i in range(len(APO_B2AR)):
    trajlist.append(APO_B2AR[i])

t = md.load(trajlist, top=PDB)
    
```

```python
Cluster = KCenters(n_clusters=16, metric=md.rmsd)
clusters = Cluster.fit(tlist)
```

```python
time_scales={}
lag_times=range(1,25)
for lag_time in lag_times:
    Msm=MarkovStateModel(lag_time=lag_time)
    Msm.fit(clusters.labels_)
    timescales=Msm.timescales_
    time_scales[lag_time]=timescales
    print('n_states=%d\tlag_time=%d\ttimescales=%s' % (16, lag_time, timescales))
```

```python
colors = list("rgbcmykr")
timescales=[0]*7
for n in range(7):
    timescales[n]=[]
    for m in range(1,21):
        timescales[n].append(time_scales[m][n])
        
for i in range(len(timescales)):
    scatter(lag_times[:20], timescales[i], color=colors[i])
title('Relaxation Timescales vs Lagtime')
xlabel('Lag Time')
ylabel('Relaxation Timescale')
#savefig('Kcenters_16clusters2.pdf')
```
